```python
import javax.swing.*;
import java.awt.*;
import java.awt.event.ActionEvent;
import java.awt.event.ActionListener;

public class Incrementador {

    private JFrame frame;
    private JLabel label;
    private JTextField textField;
    private JButton incrementarButton;
    private JButton zerarButton;

    private int valor = 0;

    public Incrementador() {
        frame = new JFrame("Incrementador");
        frame.setDefaultCloseOperation(JFrame.EXIT_ON_CLOSE);
        frame.setSize(300, 150);

        label = new JLabel("Valor: " + valor);
        label.setFont(new Font("Arial", Font.PLAIN, 20));
        label.setHorizontalAlignment(SwingConstants.CENTER);

        textField = new JTextField(10);
        textField.setHorizontalAlignment(SwingConstants.CENTER);
        textField.setEditable(false);

        incrementarButton = new JButton("Incrementar");
        incrementarButton.addActionListener(new ActionListener() {
            public void actionPerformed(ActionEvent e) {
                valor++;
                label.setText("Valor: " + valor);
            }
        });

        zerarButton = new JButton("Zerar");
        zerarButton.addActionListener(new ActionListener() {
            public void actionPerformed(ActionEvent e) {
                valor = 0;
                label.setText("Valor: " + valor);
            }
        });

        frame.setLayout(new GridLayout(3, 1));
        frame.add(label);
        frame.add(textField);
        frame.add(incrementarButton);
        frame.add(zerarButton);

        frame.setVisible(true);
    }

    public static void main(String[] args) {
        SwingUtilities.invokeLater(new Runnable() {
            public void run() {
                new Incrementador();
            }
        });
    }
}
```
